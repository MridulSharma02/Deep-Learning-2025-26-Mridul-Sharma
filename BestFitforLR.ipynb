```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
data = {
    'Hours': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Marks': [10, 20, 30, 25, 50, 60, 65, 70, 85]
}

df = pd.DataFrame(data)
print("Dataset:\n", df, "\n")
```

```python
X = df[['Hours']]   # independent variable
y = df['Marks']     # dependent variable
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
m = model.coef_[0]       # slope
c = model.intercept_      # intercept
print(f"Equation of Best Fit Line: y = {m:.2f}x + {c:.2f}\n")
```

```python
y_pred = model.predict(X)
```

```python
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
corr = np.corrcoef(y, y_pred)[0, 1]

print("Model Evaluation Metrics:")
print(f"  Mean Absolute Error (MAE): {mae:.2f}")
print(f"  Mean Squared Error (MSE): {mse:.2f}")
print(f"  Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"  R² Score: {r2:.2f}")
print(f"  Correlation Coefficient (r): {corr:.2f}\n")
```

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Best Fit Line')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.title('Simple Linear Regression: Hours vs Marks')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

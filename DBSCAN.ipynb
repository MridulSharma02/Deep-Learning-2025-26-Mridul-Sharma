```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ankit8467/dataset-for-dbscan")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv("/kaggle/input/dataset-for-dbscan/Clustering_gmm (2).csv")
```

```python
X = df[['Weight', 'Height']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)
```

```python
df['Cluster'] = labels
print(df.head())
print("\nCluster Counts:")
print(df['Cluster'].value_counts())
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Weight'], df['Height'], c=df['Cluster'], cmap='viridis')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("DBSCAN Clustering on Weight-Height Data")
plt.show()
```
